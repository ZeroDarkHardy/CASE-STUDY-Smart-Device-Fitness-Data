```python
import csv
import pandas as pd
from datetime import *
```

```python
file_to_load = "heartrate_seconds_merged.csv"

df = pd.read_csv(file_to_load, low_memory=False)
df.head(10)
```

```python
df['Time'] = df['Time'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %I:%M:%S %p'))
df.head(10)
```

```python
df.to_csv('heartrate_seconds_cleaned.csv', index=False)
```
